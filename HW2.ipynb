```python
#pip install pandas
#pip install numpy
import numpy as np
import pandas as pd
df = pd.read_excel('Реестр_выполненных_рейсов_за_период.xlsx')
df.columns = df.iloc[0]
df = df.drop(0).reset_index(drop=True)
df = df.drop(11268).reset_index(drop=True)
df = df.drop(11268).reset_index(drop=True)
df = df.drop(11268).reset_index(drop=True)
df.dropna(how='all', axis=1, inplace=True)
df.reset_index().tail()
```

**Задание 5**

---



```python
select = "Domestic" # @param ["Domestic", "International"]
topcount = df.loc[df['Направление'] == select, 'Код АП'].value_counts().reset_index()
topcount.columns = ['Код АП', 'Кол-во рейсов']
top = topcount['Кол-во рейсов'].max()
pop = topcount[topcount['Кол-во рейсов'] == top]['Код АП']
print ('\033[1m'+'Самый популярный маршрут по каждому из направлений', select, ':' +'\033[0m', ','.join(pop))
```
