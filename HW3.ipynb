**Задание 1**

```python
#pip install pandas
#pip install numpy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('Реестр_выполненных_рейсов_за_период.xlsx', header=1)
df.dropna(how='all', axis=1, inplace=True)
df.dropna(thresh=12, axis=0,inplace=True)
df.tail()
```

```python
df.columns
```

**Задание 2**

```python
fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_size_inches(10, 5)
ax1.hist(df['Вместимость']);
ax2.hist(df['Обслужено пассажиров']);
ax1.set_title('Вместимость',size=14)
ax2.set_title('Обслужено пассажиров',size=14)
fig, (ax3, ax4, ax5) = plt.subplots(1,3)
fig.set_size_inches(15, 5)
ax3.hist(df['Baggage Weight']);
ax3.set_title('Baggage Weight',size=14)
ax4.hist(df['МВМ, тонн']);
ax4.set_title('МВМ, тонн',size=14)
ax5.hist(df['Кл.3']);
ax5.set_title('Кл.3',size=14)
```
